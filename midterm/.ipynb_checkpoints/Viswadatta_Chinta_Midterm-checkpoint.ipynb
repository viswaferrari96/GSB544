1.1 What is the length in miles of the fourth longest trail in California?

```python
ca_trails = df[df['state_name'] == 'California'] #filter for California
sorted_ca_trails = ca_trails.sort_values(by='length', ascending=False) #sorting in descending
len_in_m = sorted_ca_trails.iloc[3]['length'] #4th largest trail
m_to_miles = 1 / 1609.34 #length in metres
length_in_miles = len_in_m * m_to_miles #metres to miles
length_in_miles
```

1.2 Consider an “easy” trail to be one that has an elevation gain of less than 500 meters, a distance of less than 5 kilometers, and a difficulty rating below 4.

Which California National Park has the highest average rating of its “easy” trails?

```python
easy_trails_criteria = (ca_trails['elevation_gain'] < 500) & (ca_trails['length'] < 5000) & (ca_trails['difficulty_rating'] < 4) #easy trails

easy_ca_trails = ca_trails[easy_trails_criteria] #filtering all easy trails

avg_rting_by_park = easy_ca_trails.groupby('area_name')['avg_rating'].mean() #avg rating of trails

high_avg_rating_park = avg_rting_by_park.idxmax(), avg_rting_by_park.max() #highest av rating
high_avg_rating_park

```

1.3 How many National Parks contain the word “Land” somewhere in their name? (if the word appears as part of a longer word, like “Candyland”, this counts too)

```python
landp_count = 0

for pname in df['area_name'].unique(): #looping on unique national parks
    if 'land' in pname.lower(): #checking in lowercase as well
        landp_count += 1

landp_count
```

2.2 scrape_temps function

```python
def scrape_temps(month=None):

    base_url = "https://www.extremeweatherwatch.com/us-state-averages"

    if month:
        m_url = f"{base_url}/month-{month.lower()}"  #modifying url to get the correct selection based on the selected month
    else:
        m_url = base_url

    response = requests.get(m_url)
    soup = BeautifulSoup(response.content, 'html.parser') #http request
    table = soup.find_all("table",
                  attrs={
                      "class": "bordered-table sort-table"
                  }
                  )[0]

    data = []
    for row in table.find_all('tr')[1:]:
      cells = row.find_all('td')
      if len(cells) > 0:
        city = cells[0].get_text(strip=True)
        high = cells[1].get_text(strip=True)
        low = cells[2].get_text(strip=True)
        prec = cells[3].get_text(strip=True).split()[0]
        data.append([city, high, low, prec])
    df = pd.DataFrame(data, columns=['City', 'Highest Temp', 'Lowest Temp', 'Precipitation'])
    df
    return df

scrape_temps(month="january")
```
